## 5.Outlier Detection

```python
df.head()
```

```python
from scipy import stats
z_scores = np.abs(stats.zscore(df["PassengerId"]))
outliers=df["PassengerId"][z_scores>3]
outliers
```

```python
z_scores = np.abs(stats.zscore(df["Fare"]))
outliers=df["Fare"][z_scores>3]
outliers
```

```python
Q1=df["Fare"].quantile (0.25)
Q3=df["Fare"].quantile (0.75)

IQR = Q3-Q1

lower_bound = Q1-1.5* IQR
upper_bound = Q3 + 1.5 * IQR

df_cleaned = df[(df["Fare"]> lower_bound) & (df["Fare"] < upper_bound)]

print (f"Original DataFrame size: {df.shape}") 
print (f"Cleaned DataFrame size: {df_cleaned.shape}")
df_cleaned
```
